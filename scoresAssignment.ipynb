```python
mathScore=int(input("enter your math score:"))
scienceScore=int(input("enter your science score:"))
englishScore=int(input("enter your english score:"))
average=(mathScore+scienceScore+englishScore)/3
if average>=90:
    print("Excellent! You got an A")
elif 80<=average<90:
    print("Very Good! You got a B")
elif 70<=average<80:
    print("Good! You got a C")
elif 60<=average<70:
    print("Needs Improvement! You got a D")
elif average<60:
    print("Fail! You got an F")
elif average==100:
    print("Wow!You got a perfect score")
elif average==0:
    print("Oh No!You failed everything")
else:
    print("Invalid Input.Try again!")
```
